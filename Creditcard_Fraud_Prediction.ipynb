```python
## Feature SCaling(Values are not in the same scaling),(We put numerical values in one stard scale)
from sklearn.preprocessing import StandardScaler
```

```python
st= StandardScaler()
df['Amount'] = st.fit_transform(pd.DataFrame(df['Amount']))
```

```python
df.head()
```

```python
##Droping Time column
df = df.drop(columns='Time')
```

```python
df.head()
```

```python
##Check for duplicates
df.duplicated().any()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.duplicated().any()
```

```python
df.shape
```

```python
df['Class'].value_counts()
```

```python
### Data is IMBALANCED
```

```python
##independent an ddependent variable
X = df.drop(columns='Class')
Y = df['Class']
```

```python
## Split data into training and testing sets
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y, test_size=0.2, random_state=42)
```

```python
X_train.shape,X_test.shape,Y_train.shape,Y_test.shape
```

```python
##Handling IMbalanced datasets
```

```python
#Undersampling
```

```python
normal=df[df['Class']==0]
fraud=df[df['Class']==1]
```

```python
normal.shape
```

```python
fraud.shape
```

```python
normal_sample=normal.sample(n=473)
```

```python
normal_sample.shape
```

```python
new_data= pd.concat([normal_sample,fraud],ignore_index=True)
```

```python
new_data.shape
```

```python
##independent an ddependent variable
X = new_data.drop(columns='Class')
Y = new_data['Class']
```

```python
 X_train,X_test,Y_train,Y_test = train_test_split(X,Y, test_size=0.2, random_state=42)
```

```python
 X_train.shape,X_test.shape,Y_train.shape,Y_test.shape
```

```python
##Logistic Regression
from sklearn.linear_model import LogisticRegression
```

```python
log = LogisticRegression()
log.fit(X_train,Y_train)
```

```python
y_pred1=log.predict(X_test)
```

```python
y_pred1
```

```python
## Check the accuracy of the model
from sklearn.metrics import accuracy_score
```

```python
##accuracy of the model
a = accuracy_score(Y_test,y_pred1)
```

```python
a
```

```python
from sklearn.metrics import precision_score,recall_score,f1_score
```
